```python
#get average of dark images to create offset
import os
import cv2
import shutil
import numpy as np

#change this to access correct folder 
directory = 'C:/Users/Wendy/Documents/GitHub/MBImager/build-ImagerGUI-Desktop_Qt_5_9_1_MinGW_32bit-Debug/MBImagerGUI/Dark3'
img = np.zeros((80,60))
for subdir in os.listdir(directory):
    print subdir
    subdirpath = os.path.join(directory,subdir)
    cnt = 0
    for filename in os.listdir(subdirpath):
        #print filename
        img+= cv2.imread(os.path.join(subdirpath,filename),cv2.IMREAD_GRAYSCALE )
        cnt +=1 
    img = img/cnt
    cv2.imwrite(subdirpath + '.png', img)
    print(img)
```

```python
stdDict = {}
meanDict = {}
varDict = {}
lightdirectory = 'C:/Users/Wendy/Documents/GitHub/MBImager/build-ImagerGUI-Desktop_Qt_5_9_1_MinGW_32bit-Debug/MBImagerGUI/Light3'
darkdirectory = 'C:/Users/Wendy/Documents/GitHub/MBImager/build-ImagerGUI-Desktop_Qt_5_9_1_MinGW_32bit-Debug/MBImagerGUI/Dark3'

for subdir in os.listdir(lightdirectory):
    print subdir
    data = np.zeros((100,80,60))
    subdirpath = os.path.join(lightdirectory,subdir)
    cnt = 0
    for filename in os.listdir(subdirpath):
        #print filename
        data[cnt] = cv2.imread(os.path.join(subdirpath,filename),cv2.IMREAD_GRAYSCALE ) - cv2.imread(os.path.join(darkdirectory,subdir)+'.png',cv2.IMREAD_GRAYSCALE ) 
        cnt +=1
    pixstd = np.zeros((80,60))
    pixmean = np.zeros((80,60))
    pixvar = np.zeros((80,60))
    for row in range(80):
        for col in range(60):
            pixstd[row,col] = np.std(data[:,row,col])
            pixmean[row,col] = np.mean(data[:,row,col])
            pixvar[row,col] = np.var(data[:,row,col])
    stdDict[subdir] = pixstd
    meanDict[subdir] = pixmean
    varDict[subdir] = pixvar

```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
logmean= np.zeros((80,60,len(stdDict.keys())))
lognoise= np.zeros((80,60,len(stdDict.keys())))
i = 0
for key in sorted(stdDict.keys()):
    for row in range(80):
        for col in range(60):

            logmean[row,col,i] = math.log(meanDict[key][row, col],10)
            try:
                lognoise[row,col, i] = math.log(stdDict[key][row,col],10)
            except:
                lognoise[row,col, i] = 0
            
    
    i +=1
for row in range(80):
    for col in range(60):
        plt.plot(logmean[row, col], lognoise[row,col], 'o')
plt.show()
```
